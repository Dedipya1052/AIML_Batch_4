```python
import matplotlib.pyplot as plt

# Generate dataset
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(age, height, color='blue', label='Height')
plt.scatter(age, weight, color='red', label='Weight')
plt.title('Health Data Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Measurements')
plt.legend()
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, bar_width, color='blue', label='Height')
plt.bar([i + bar_width for i in index], weight, bar_width, color='red', label='Weight')
plt.xlabel('People')
plt.ylabel('Measurements')
plt.title('Health Data Bar Chart')
plt.xticks([i + bar_width / 2 for i in index], people)
plt.legend()
plt.grid(True)
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
plt.hist([height, weight], bins=5, color=['blue', 'red'], label=['Height', 'Weight'])
plt.xlabel('Measurements')
plt.ylabel('Frequency')
plt.title('Health Data Histogram')
plt.legend()
plt.grid(True)
plt.show()
```
